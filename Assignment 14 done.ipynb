```python
df['Avg_MPG'] = (df['city mpg'] + df['highway MPG']) / 2
avg_values = df.groupby('Year').agg({'Engine HP': 'mean','Avg_MPG': 'mean'}).round() 
avg_values
```

```python
total_cars = len(df)
size_counts = df['Vehicle Size'].value_counts()
percentage = (size_counts[['Compact','Midsize','Large']] / total_cars) * 100
percentage = percentage.round(2)
percentage
```

```python
plt.boxplot(df['Engine HP'].dropna())
plt.title("Boxplot of Engine HP")
plt.ylabel("Engine Horsepower")
plt.show()
```

```python
cylinder = df['Engine Cylinders'].dropna()

plt.boxplot(cylinder)
plt.title("Boxplot of Engine Cylinders")
plt.ylabel("Number of Cylinders")
plt.show()
```

```python
hp = df['Engine HP'].dropna()

Q1 = hp.quantile(0.25)
Q3 = hp.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = hp[(hp < lower_bound) | (hp > upper_bound)]
num_outliers = outliers.count()

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Number of Outliers: {num_outliers}")
```
